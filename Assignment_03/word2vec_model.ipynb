```python
import pandas as pd
```

```python
raw_df = pd.read_json('Tools_and_Home_Improvement_5.json', lines=True)
```

```python
raw_df
```

```python
raw_corpus = u"".join(raw_df['reviewText']+" ")
```

```python
import spacy
from gensim.models.phrases import Phrases, Phraser
from spacy.pipeline import Pipe
```

```python
nlp = spacy.load('en')
```

```python
%%time
token_text = []

for doc in nlp.pipe(raw_df.reviewText):
    for sent in doc.sents:
        token_text.append([t.lower_ for t in sent if not t.is_punct])
```

```python
token_text[0]
```

```python
import gensim
```

```python
%%time

model = gensim.models.word2vec.Word2Vec(token_text, seed = 1,min_count = 5, size=100, workers=8)
```

```python
model.wv.most_similar('flashlight')
```

```python
len(pd.Series(list(model.wv.vocab)))
```

```python
model.save("Tools_and_Home_word2vec.w2v")
```
